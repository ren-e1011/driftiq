```python
def get_spikes_by_direc(imix, refrac_pd,threshold,nsteps,npause):
    vec_counts_ = {"N":[],"S":[],"W":[],"E":[],"NW":[],"NE":[],"SW":[],"SE":[]}

    vec_counts_clean = {"N":[],"S":[],"W":[],"E":[],"NW":[],"NE":[],"SW":[],"SE":[]}

    for vec in vec_counts_.keys():
        events, counts, traj = im2linewalk2events(i=imix,vec=vec,refrac_pd=refrac_pd, thres=threshold, nsteps=nsteps,npause=npause)
        vec_counts_[vec] = counts
        
        cutevents, cutcounts = cutEdges(events,traj)
        # cutcounts = [len(cutevent) for cutevent in cutevents]
        vec_counts_clean[vec] = cutcounts

    return vec_counts_, vec_counts_clean
```

### No RP, high threshold

```python
ix_raw = {}
ix_cut = {}
for ix in imix_list:
    rawspikecount, cleanspikecount = get_spikes_by_direc(imix=ix,refrac_pd=refrac_pd_list['low'],threshold=threshold['high'],nsteps=60,npause=40)
    ix_raw[ix] = rawspikecount
    ix_cut[ix] = cleanspikecount
```

```python
for ix in ix_raw.keys():
    cut = ""
    pd.DataFrame(ix_raw[ix]).plot(title=f"{cut} Im {ix} spike counts by direction {refrac_pd_list['low']} RP {threshold['high']} threshold")
```
